```python
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
#import all 3 csv
census_data = "census_data.csv"
state_vacc = "state_vacc.csv"
state_abrev = "state_abrev.csv"
```

```python
#turn the csv into dataframes
census_df = pd.read_csv(census_data)
state_vacc_df = pd.read_csv(state_vacc)
state_abrev_df = pd.read_csv(state_abrev)
```

```python
#double check the dataframes are correct
#census_df.head()
#state_vacc_df.head()
#state_abrev_df.head()
```

```python
state_vacc_df = state_vacc_df.rename(columns = {'state': 'state abrev'})
```

```python
state_abrev_df = state_abrev_df.rename(columns = {'Abrev': 'state abrev'})
```

```python
#merge census withs state_abrev
census = pd.merge(census_df, state_abrev_df, on = "State", how ="left" )
census.head()
```

```python
#merge state_vacc withs state_abrev
state_vacc = pd.merge(state_vacc_df, state_abrev_df, on = "state abrev", how ="left" )
state_vacc.head()
```

```python
#merge census and state df
econ_vacc = pd.merge(census, state_vacc, on = "State", how ="inner" )
econ_vacc.head()
```

```python
econ_vacc = econ_vacc[['State', 'state abrev_y', 'Total Population', 'Total Labor Force', 'Total Unemployed', 'Total Poverty', 'State Number', 'Poverty Rate', 'Unemployment Rate', 'total vaccines distributed', 'total vaccines administered', 'total vaccinated', 'pct vaccinated', 'boosters administered', 'additional_doses_vax_pct']]
```

```python
econ_vacc = econ_vacc.rename(columns = {'state abrev_y': 'State Abrev',
                                       'total vaccines distributed': 'Total Vaccines Distributed', 
                                        'total vaccines administered': 'Total Vaccines Administered',
                                        'total vaccinated': 'Total Vaccinated', 
                                        'pct vaccinated': 'Percent Vaccinated', 
                                        'boosters administered': 'Boosters Administered', 
                                        'additional_doses_vax_pct': 'Additional Doses Pct'})
```

```python
econ_vacc.head()
```

```python
conn = 'mongodb://127.0.0.1:27017'
```

```python
client = pymongo.MongoClient(conn)
```

```python
db = client.vaccinationDB
```

```python
statevacc = db.statevacc
```

```python
econ_vacc_dict = econ_vacc.to_dict("records")
```

```python
statevacc.insert_many(econ_vacc_dict)
```

```python
results = statevacc.find()
for result in results:
    print(result)
```
